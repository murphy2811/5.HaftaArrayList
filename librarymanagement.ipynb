```python
class Library:


    def __init__(self):

        self.file = open("books.txt", "a+")


    def __del__(self):

        self.file.close()


    def list_books(self):

        self.file.seek(0)
        file_contents= self.file.read()
        list = file_contents.splitlines()

        for book_info in list:

                line = book_info.split(',')

                print(f"Book name:{line[0]},Author:{line[1]}")

    def add_books(self):

        book_name = input("Book name:")
        author = input("Author:")

        x=False
        y=False
        deneme = 0

        while True:

            pages =input("Enter the number of pages: ")
            if pages.isdigit():
               break
            print("Invalid input. Please enter a positive integer for the number of pages.")

        while True:
            year = input("Enter the year: ")
            if year.isdigit():
               break
            print("Invalid input. Please enter a valid year.")




        info=f"{book_name},{author},{year},{pages}"

        self.file.write(info)
        print(f"{book_name},has been added to the library")

    def remove_books(self):
        removing = input("Enter the title of the book to remove: ")

        self.file.seek(0)
        file_contents = self.file.read()
        list = file_contents.splitlines()

        founded=False
        for index in range(len(list)):
          book_info = list[index]
          line = book_info.split(',')
          if removing == line[0]:
                index_to_remove = index
                founded=True
                break

        if founded:

            del list[index_to_remove]


            self.file.seek(0)
            self.file.truncate()


            for book_info in list:
                self.file.write(book_info + '\n')

            print(f"The book '{removing}' has been removed.")

        else:
            print(f"The book '{removing}' was not found.")




while True :
    lib = Library()
    print("*** MENU ***")
    print("1) List Books")
    print("2) Add Book")
    print("3) Remove Book")

    option = input("Enter your choice (1-3, or 'q' to quit): ")


    if option == '1':
        lib.list_books()
        del lib
    elif option == '2':
        lib.add_books()
        del lib
    elif option == '3':
        lib.remove_books()
        del lib
    elif option.lower() == 'q':
        del lib
        break
    else:
        print("Invalid option. Please enter 1, 2, 3, or 'q' to quit.")
```
